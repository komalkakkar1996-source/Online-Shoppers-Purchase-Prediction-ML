Five different machine learning experiments were performed to identify the most suitable model for predicting online shopper purchase intention. Each experiment was evaluated using Accuracy, Precision, Recall, F1-Score, ROC-AUC, and Overfitting. Since the dataset is imbalanced, model selection was based on multiple evaluation metrics rather than accuracy alone.

### Based on overall predictive performance, the models are ranked as follows:
**RANK 1. Exp4 – Random Forest (Best Overall)**
- Highest F1 (0.69)
- Excellent ROC-AUC (0.9285)
- Good Precision + Good Recall
- Strongest overall predictive performance

**RANK 2. Exp2 – Logistic Regression + Log1p**
- Best Logistic Regression
- Highest Accuracy (0.9029)
- Interpretable and balanced

**RANK 3. Exp3 – Logistic Regression + Log1p + Class Weight**
- Best when Recall is the priority
- Highest ROC-AUC (0.9293)
- Recall improved from 0.61 to 0.81, meaning the model detected more purchasing customers. However, this improvement came with a reduction in Accuracy and Precision, resulting in more false positive predictions.

- This model is suitable when missing a potential buyer is more costly than contacting extra non-buyers.

**RANK 4. Exp4 – Tuned Decision Tree**
- High Recall (0.81) but model still produced the lowest Accuracy (85%), lowest Precision (0.52), and lower F1-score (0.63). Therefore, it was outperformed by both Logistic Regression and Random Forest.

**RANK 5. Exp1 – Baseline Logistic Regression** 
- This model serves as a benchmark. While it achieved the highest Precision (0.77), its Recall (0.42) was very low, meaning it missed a large number of actual buyers.

# Conculsion

Although Logistic Regression with Log1p Transformation achieved the highest accuracy (90.29%), accuracy alone was not considered sufficient because the dataset is imbalanced. Greater emphasis was placed on Recall, F1-Score, ROC-AUC, and model generalization.

"I would choose the tuned Random Forest because it provides the best overall balance between Accuracy, Precision, Recall, F1-Score, ROC-AUC, and Generalization.

It has the highest F1-score (0.69), indicating the best trade-off between precision and recall, and excellent ROC-AUC (0.9286), showing excellent class separation. 

Since customer purchase prediction requires both identifying buyers accurately and minimizing missed buyers, Random Forest is the most suitable production model."
